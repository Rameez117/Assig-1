```python
#Even number (sum )
import numpy as np
def sum(np):
    total = 0
    for x in np:
        total += x
    return total

eve_numbers=(2, 4, 6, 8,10,12,14,16)
print("Sum of Even Numbers is equals to : ",sum(eve_numbers))
```

```python
# Display image in different colors
import imageio
import matplotlib.pyplot as plt
import numpy as np
im = imageio.imread('C:/Users/CHAUDARY/Desktop/Dataset/car.jpg')
im_r = np.zeros(np.shape(im))
im_r[:, :, 0] = im[:, :, 0]
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].imshow(im[:, :, 0], cmap='Reds_r')
axs[0, 1].imshow(im_r.astype(int))
axs[1, 0].imshow(im[:, :, 0])
axs[1, 1].imshow(im[:, :, 0], cmap='Greys_r')
axs[0, 0].set_title('red ')
axs[0, 1].set_title('dark red ')
axs[1, 0].set_title('green ')
axs[1, 1].set_title('gray')
plt.tight_layout()
```
